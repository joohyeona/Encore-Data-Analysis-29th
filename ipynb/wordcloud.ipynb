```python
import pandas as pd
```

```python
survey_raw_df = pd.read_csv('datas/survey_results_public.csv', index_col='ResponseId')
```

```python
dev_type = survey_raw_df.DevType.copy()
```

```python
dev_type.dropna(inplace=True)
```

```python
#;를 공백으로, ,를 공백으로 변경
dev_type = dev_type.str.replace(';', ' ').str.replace(',', ' ')
```

```python
words = ' '.join(dev_type)
#하나의 문자열로 합치기
```

```python
import matplotlib.pyplot as plt

#한글이 포함되었으면 한글 폰트 지정해야 함
#font_path=['폰트 경로']
#max_font_size = 60 최대 글자 크기 지정

wc = WordCloud(collocation_threshold=int(1e6),
               width=800, height=400,
               background_color='lightgrey').generate(words)

plt.figure(figsize=(16, 8))

#x, y축을 그리지 않기
plt.axis('off')

#grid lines 그리지 않기
plt.grid(False)
plt.imshow(wc)

plt.show()
```

```python
wc.to_file('DevType.png')
#image 파일로 저장
```
