```python
#틱 설정하기
#틱(tick): 플롯이나 차트에서 축상의 위치 표시 지점

import numpy as np
X = np.linspace(-np.pi, np.pi, 256)
#      linspace: 등간격으로 범위 분할
C = np.cos(X)

plt.title('x축과 y축의 tick label 설정')
plt.plot(X, C)
plt.xticks([-np.pi, -np.pi /2, 0, np.pi/2, np.pi])
plt.yticks([-1, 0, 1])
plt.show()
```

```python
"""
틱 라벨 문자열에 수학 문자식 넣기(Mathematical expressions)
https://matplotlib.org/3.1.1/tutorials/text/mathtext.html
"""

X = np.linspace(-np.pi, np.pi, 256)
C = np.cos(X)

plt.title('x축과 y축의 tick label 설정')
plt.plot(X, C)
plt.xticks([-np.pi, -np.pi /2, 0, np.pi/2, np.pi], 
           [r'$-\pi$', r'$-\pi/2$', r'$0$', r'$+\pi/2$', r'$+\pi$'])
plt.yticks([-1, 0, 1], ['Low', 'Zero', 'High'])
plt.show()
```

연습문제

- title: 섭씨온도와 화씨온도
- xlabel : 화씨 온도
- ylabel : 섭씨 온도
- x 값은 화씨로 32 이상~ 100미만의 정수,

```python
x1 = np.linspace(0.0, 5.0)
x2 = np.linspace(0.0, 2.0)
y1 = np.cos(2 * np.pi * x1) * np.exp(-x1)
y2 = np.cos(2 * np.pi * x2)

ax1 = plt.subplot(2, 1, 1)  #첫번째 거부터 접근!
plt.plot(x1, y1, 'yo-')
plt.title('A tale of 2 subplots')
plt.ylabel('Damped oscillation')

ax2 = plt.subplot(2, 1, 2)
plt.plot(x2, y2, 'r.-')
plt.xlabel('time (s)')
plt.ylabel('Undamped')

plt.tight_layout()
plt.show()
```

### 파이 차트

```python
labels = ['월세', '전세', '자가', '기타']
sizes = [15, 30, 45, 10]
colors = ['yellowgreen', 'gold', 'lightskyblue', 'lightcoral']
explode = (0, 0, 0.1, 0)

plt.title('Pie Chart')
plt.pie(sizes, explode = explode, labels=labels, colors=colors,
        autopct='%1.1f%%', shadow=True, startangle=90)
plt.axis('equal') #원의 형태를 유지, 현재는 없어도 큰 달라짐 없음
plt.show()
```

```python
import seaborn as sns
titanic = sns.load_dataset('titanic')

bins = [0, 20, 30, 50, 70, 100]
ages = titanic.age.value_counts(bins=bins, dropna=True, sort=False)

labels = ["미성년자", "청년", "장년", "중년", "노년"]
explode = (0.1, 0, 0, 0, 0)
colors = ["yellowgreen", "gold", "lightskyblue", "lightcoral", "orange"]

plt.pie(ages, explode = explode, labels=labels, colors=colors,
        autopct='%1.1f%%', shadow=True, startangle=90)

plt.show()
```

```python
import pandas as pd

bins = [0, 20, 30, 50, 70, 100]
ages = titanic.age.value_counts(bins=bins, dropna=True, sort=False)

labels = ["미성년자", "청년", "장년", "중년", "노년"]
explode = (0.1, 0, 0, 0, 0)
colors = ["yellowgreen", "gold", "lightskyblue", "lightcoral", "orange"]

age_df = pd.DataFrame(data=ages.values, index=labels)

plt.pie(ages, explode = explode, labels=age_df.index, colors=colors,
        autopct='%1.1f%%', shadow=True, startangle=90)
plt.axis('equal')
plt.show()
```
